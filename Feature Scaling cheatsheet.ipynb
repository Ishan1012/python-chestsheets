# **Feature Scaling**: Definition of Feature Scaling, Min-Max Scaling

```python
import numpy as np
from sklearn.preprocessing import MaxAbsScaler, MinMaxScaler
```

```python
data =  np.array([[-100],[50],[-20],[80]])
```

```python
scaler = MaxAbsScaler()
```

```python
scaled_data = scaler.fit_transform(data)
```

```python
print("Original data: \n",data)
print("Scaled data: \n",scaled_data)
```

```python
scaler = MinMaxScaler()
```

```python
scaled_data = scaler.fit_transform(data)
```

```python
print("Original data: \n",data)
print("Scaled data: \n",scaled_data)
```

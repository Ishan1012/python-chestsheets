# Confusion Matrix

- A Confusion Matrix shows how many predictions your model got right or wrong.
- Components of Confusion Matrix:
    1. **True Positive (TP)**: Model predicted positive and it is actually positive.
    2. **False Positive (FP)**: Model predicted positive and it is actually negative.
    3. **False Negative (FN)**: Model predicted negative and it is actually positive.
    4. **True Negative (TN)**: Model predicted negative and it is actually negative.
- It is called confusion matrix because it shows where model gets confused, i.e., which class it is mixing up (FP and FN).

\$$Accuracy = \frac{TP+TN}{TP+TN+FP+FN}$\$\
\$$Precision = \frac{TP}{TP+FP}$\$\
\$$Recall = \frac{TP}{TP+FN}$\$
